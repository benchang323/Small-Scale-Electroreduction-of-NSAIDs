# AP Research Data Processing Notebook 1: Pre-Experiment
The following codes in this Jupyter notebook (E0.ipynb) contains the data analysis portion of Benjamin Chang's AP Research project on *Optimal Conditions For Small-Scale Electroreduction of Common Non-Steroidal Anti-Inflammatory Drugs*. All charts, graphs, and figures are produced from the raw data in .csv files and manipulated by the following codes.

## Initializing Libraries and Style

```python
#Import Libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot
```

```python
#Set font format
font = {
    'family' : 'georgia',
    'color': 'black',
    'weight': 'normal',
    'size': 11
}
```

## Salicylic Acid

```python
#Load CSV data
dfSA = pd.read_csv('Raw Data/E0-SA.csv', delimiter = ',')
```

```python
#Create new column for average absorbance
col = dfSA.loc[:, 'Absorbance 1 (Au)':'Absorbance 3 (Au)']
dfSA['Absorbance (Au)'] = col.mean(axis=1)

#Create new column for standard deviation
col = dfSA.loc[:, 'Absorbance 1 (Au)':'Absorbance 3 (Au)']
dfSA['Standard Deviation'] = col.std(axis=1)
```

```python
#Plot Absorbance vs. Wavelength line black
pyplot.plot(dfSA['Wavelength (nm)'], dfSA['Absorbance (Au)'], color = 'black')
pyplot.xlabel('Wavelength (nm)', fontdict = font)
pyplot.ylabel('Absorbance (Au)', fontdict = font)

#Remove the top and right border of the chart
pyplot.gca().spines['top'].set_visible(False)
pyplot.gca().spines['right'].set_visible(False)

#Draw a vertical line at x = 800
pyplot.axvline(x=800, color='k', linestyle='--')

pyplot.savefig('Graphs/E0-SA.jpg', dpi = 500)
pyplot.show()
```

```python
#Printing data output for all (300-1105nm)
print("Maximum absorbance in all (Au): ", dfSA['Absorbance (Au)'].max())
print("Optimal wavelength in all (nm): ", dfSA[dfSA['Absorbance (Au)'] == dfSA['Absorbance (Au)'].max()]['Wavelength (nm)'].item())

#Printing data output in UV-VIS range (300-800nm)
print("\nMaximum absorbance in UV-VIS (Au): ", dfSA['Absorbance (Au)'][800 >= dfSA['Wavelength (nm)']].max())
print("Optimal wavelength in UV-VIS (nm): ", dfSA[dfSA['Absorbance (Au)'] == dfSA['Absorbance (Au)'][800 >= dfSA['Wavelength (nm)']].max()]['Wavelength (nm)'].item())

#Create new dataframe to save maximum absorbance wavelength and value for UV-VIS range
dfUVVIS = pd.DataFrame(columns = ['Wavelength (nm)', 'Absorbance (Au)'])

#Add data
dfUVVIS.loc[0] = [dfSA[dfSA['Absorbance (Au)'] == dfSA['Absorbance (Au)'][800 >= dfSA['Wavelength (nm)']].max()]['Wavelength (nm)'].item(), dfSA[dfSA['Absorbance (Au)'] == dfSA['Absorbance (Au)'][800 >= dfSA['Wavelength (nm)']].max()]['Absorbance (Au)'].item()]
```
